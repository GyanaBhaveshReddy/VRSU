```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
image_path=list(uploaded.keys())[0]
image=cv2.imread(image_path,0)
img_1=cv2.imread(image_path)
img_rgb=cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB)
if image is None:
  raise ValueError("Failed to load image!")
threshold_value, binary_img=cv2.threshold(image,100,255,cv2.THRESH_BINARY)
print("Optimal Threshold Value:",threshold_value)
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')
plt.subplot(1,3,2)
plt.title("Histogram")
plt.hist(image.ravel(),256)
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.subplot(1,3,3)
plt.title("Otsu Binarized Image")
plt.imshow(binary_img,cmap='gray')
plt.axis('off')
plt.show()


#
```
