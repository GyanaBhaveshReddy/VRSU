```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/i1.jpg", 0)
h = cv2.equalizeHist(img)

plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Equalized")
plt.imshow(h, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Histogram")
plt.hist(h.ravel(), 256)

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/i3.jpg", 0)
c = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
h = c.apply(img)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("CLAHE")
plt.imshow(h, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/i2.jpg")
ycc = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
y, cr, cb = cv2.split(ycc)
c = cv2.createCLAHE(clipLimit=2.5, tileGridSize=(8,8))
yy = c.apply(y)
m = cv2.merge((yy, cr, cb))
r = cv2.cvtColor(m, cv2.COLOR_YCrCb2BGR)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,2,2)
plt.title("CLAHE")
plt.imshow(cv2.cvtColor(r, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

i1 = cv2.imread("/content/i4.webp")
i2 = cv2.imread("/content/i5.avif")
i3 = cv2.imread("/content/i6.avif")

if i1 is None or i2 is None or i3 is None:
    print("error")
    exit()

r1 = cv2.cvtColor(i1, cv2.COLOR_BGR2RGB)
r2 = cv2.cvtColor(i2, cv2.COLOR_BGR2RGB)
r3 = cv2.cvtColor(i3, cv2.COLOR_BGR2RGB)

b1 = cv2.blur(i1, (5,5))
b2 = cv2.GaussianBlur(i2, (5,5), 0)
b3 = cv2.medianBlur(i3, 5)

bb1 = cv2.cvtColor(b1, cv2.COLOR_BGR2RGB)
bb2 = cv2.cvtColor(b2, cv2.COLOR_BGR2RGB)
bb3 = cv2.cvtColor(b3, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(14,10))

plt.subplot(3,2,1)
plt.imshow(r1)
plt.axis('off')

plt.subplot(3,2,2)
plt.imshow(bb1)
plt.axis('off')

plt.subplot(3,2,3)
plt.imshow(r2)
plt.axis('off')

plt.subplot(3,2,4)
plt.imshow(bb2)
plt.axis('off')

plt.subplot(3,2,5)
plt.imshow(r3)
plt.axis('off')

plt.subplot(3,2,6)
plt.imshow(bb3)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("/content/i7.jpg", 0)
k = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
s = cv2.filter2D(img, -1, k)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(s, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread("/content/i8.jpg",0)
sx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
sy=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
s=cv2.magnitude(sx,sy)
c=cv2.Canny(img,100,200)

plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(s,cmap='gray')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(c,cmap='gray')
plt.axis('off')

plt.show()
```
